# fenestr — Demo Notebook

> **fenestr** (from Latin *fenestra*, meaning *window*) renders YouTube videos and websites
> directly inside Jupyter Notebook cells — a window into the web without leaving your notebook.

---

````python
import fenestr

print(f"fenestr version: {fenestr.__version__}")
````

---

## 1. Render a YouTube Video

Pass any YouTube URL to `fenestr.show()` or `fenestr.youtube()`.
fenestr automatically extracts the video ID and builds the embed URL.

````python
# Auto-dispatch: fenestr detects the YouTube URL
fenestr.show("https://www.youtube.com/watch?v=dQw4w9WgXcQ")
````

````python
# Explicit YouTube call with custom dimensions
fenestr.youtube(
    "https://youtu.be/dQw4w9WgXcQ",
    width=1280,
    height=720,
)
````

---

## 2. Render a YouTube Playlist

Use `render_youtube_playlist()` with a playlist ID
(the `list=` value from a YouTube playlist URL).

````python
from fenestr.youtube import render_youtube_playlist

# Replace with any public playlist ID
render_youtube_playlist("PLbpi6ZahtOH6Ar_3GPy3workGfpCBF7FG", width=900, height=500)
````

---

## 3. Render a Website

Pass any `https://` URL to embed a live website.
Here we use `docs.python.org` — a site that allows iframe embedding.

````python
fenestr.show("https://docs.python.org/3/", width=1100, height=700)
````

````python
# Explicit site call
fenestr.site("https://en.wikipedia.org/wiki/Fenestra", width=1000, height=600)
````

---

## 4. Render Raw HTML

Render locally-constructed HTML directly in the cell output —
no URL needed.

````python
fenestr.html("""
<div style="font-family: sans-serif; padding: 24px;
            border-left: 4px solid steelblue; background: #f0f4f8;">
  <h2 style="color: steelblue; margin-top: 0;">Hello from fenestr!</h2>
  <p>This is raw HTML rendered inline by
     <strong>fenestr.html()</strong>.</p>
  <ul>
    <li>No URL required</li>
    <li>Useful for custom widgets, reports, and previews</li>
  </ul>
</div>
""")
````

---

## 5. Class-Based API

Use `fenestr.Fenestr` when you want persistent width/height defaults
across multiple calls in the same notebook.

````python
f = fenestr.Fenestr(default_width=1000, default_height=600)

# All calls below use width=1000, height=600 unless overridden
f.show("https://docs.python.org/3/")
````

````python
# Override dimensions for a single call
f.youtube("https://youtu.be/dQw4w9WgXcQ", width=800, height=450)
````

---

## 6. Error Handling

fenestr raises typed exceptions you can catch and handle.

````python
from fenestr import InvalidURLError

try:
    fenestr.site("http://example.com")  # http:// not allowed
except InvalidURLError as exc:
    print(f"Caught InvalidURLError: {exc}")
````
